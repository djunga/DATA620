```python
import networkx as nx
```

```python
%matplotlib inline

g = nx.Graph()

g.add_nodes_from(["John","Jane","Joe","Mike","Dave","Lisa","Nate","Tim","Sarah"])
g.add_edges_from([("John","Nate"),("John","Mike"),("John","Dave"),
                  ("Nate","Lisa"),("Nate","Tim"),("Nate","Sarah"),
                  ("Lisa","Mike"), ("Lisa","John"),
                  ("Jane","Lisa"),("Jane","Tim"),
                  ("Joe","Jane"),("Joe","Mike"),("Joe","Dave"),
                  ("Mike","Jane"),("Mike","Nate"),
                  ("Dave","Mike"),("Dave","Lisa"),
                  ("Tim","Lisa"),("Tim","John"),
                  ("Sarah","Tim")
                  ])

nx.draw(g, with_labels=True)
```
